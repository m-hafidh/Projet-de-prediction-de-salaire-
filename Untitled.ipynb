Preparations de données 

```python
# selectionner les colonnes  "YearsExperiences" comme variable d'entrée
# et le colonne "Salary" comme variable cible

x = Data [["YearsExperience"]]
y = Data [["Salary"]]
```

```python
#Mettre les points sur un graphique 
plt.scatter(x,y)
```

```python
# Separer les données (entrainement et test)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
# Omporter la librairie de la  regression lineaire 
from sklearn.linear_model import LinearRegression
regressor =LinearRegression() # instantiation (c'est de creer)
```

```python
regressor.fit(x_train, y_train) # €ntrainement = produire l'equation lineaire 
```

```python
# Retrouver les coefficients et l'ordonnées à l'origine
print (regressor.coef_)
print (regressor.intercept_)
```

```python
# Ajouter surle graphique la droite
ordonne = np.linspace(0,15,1000)
plt.scatter(x,y ) # afficher les points  (x et y)
plt.plot(ordonne, regressor.coef_[0]*ordonne+regressor.intercept_, color="r") # tracer la ligne 
plt.xlabel ("Years of experience") # Renommer le coté experience 
plt.ylabel("Salary") # Renommer le coté salaire

```

                       Maintenant on va pasdser à la prediction 

```python
#Predictions surla base detest 
y_predict = regressor.predict(x_test)
```

```python
#Metrique
from sklearn import metrics
print ("MAE",metrics.mean_absolute_error(y_test, y_predict))
print ("MSE", metrics.mean_squard_error(y_test, y_predict))
print("RMSE",np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
print(("R²", metrics.r2_score(y_test, y_predict))

```
