                                                       Preparations de données 

```python
# selectionner les colonnes  "YearsExperiences" comme variable d'entrée
# et la colonne "Salary" comme variable cible

x = df [["YearsExperience"]]
y = df [["Salary"]]
```

```python
#Mettre les points sur un graphique 
plt.scatter(x,y)
```

```python
pip install -U scikit-learn scipy matplotlib #installer le package "sklearn"
```

```python
# Separer les données (entrainement et test)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
# APPLICATION DU MODÈLE DE RÉGRESSION LINÉAIRE
from sklearn.linear_model import LinearRegression
regressor =LinearRegression() # instantiation (c'est de creer)
regressor.fit(x_train, y_train) # €ntrainement = produire l'equation lineaire 
```

```python
# Retrouver les coefficients et l'ordonnées à l'origine
print (regressor.coef_)
print (regressor.intercept_)
```

                      VISUALISATION DE L'ENSEMBLE D'ENTRAÎNEMENT

```python
# Ajouter surle graphique la droite
#ordonne = np.linspace(0,15,1000)
#plt.scatter(x,y ) # afficher les points  (x et y)
#plt.plot(ordonne, regressor.coef_[0]*ordonne+regressor.intercept_, color="r") # tracer la ligne



plt.scatter(x_train, y_train, color ='red')# la couleur de points
plt.plot(x_train, regressor.predict(x_train), color ='blue') # tracer et donner la couleur 
plt.title("Salaire vs Experience ( entrainnement de test)")
plt.xlabel ("Années d'experience ") # Renommer le coté experience 
plt.ylabel("Salaire") # Renommer le coté salaire
plt.show()

```

             VISUALISATION DE L'ENSEMBLE DE TEST

```python
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'blue')
plt.title("Salaire vs Experience ( ensemble de test)")
plt.xlabel ("Années d'experience ") 
plt.ylabel("Salaire") 
plt.show()
```

                       Maintenant on va pasdser à la prediction 

```python
#Predictions surla base de test 
y_predict = regressor.predict(x_test)
print (y_predict)
```

```python
#Metrique
from sklearn import metrics
print ('MAE :' , metrics.mean_absolute_error(y_test, y_predict))
print ('MSE :' ,  metrics.mean_squared_error(y_test, y_predict))
print('RMSE :' , np.sqrt(metrics.mean_squared_error(y_test, y_predict)))
print('R² :' ,  metrics.r2_score(y_test, y_predict))

```
